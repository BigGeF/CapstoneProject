```python
#  Is there any connection between different categories in terms of one category directly correlating to another (E.G. if someone has a high Mobility score, does that mean they also have a high Goods or Food score?)
```

```python
# Removed column ' Carbon Footprint % ', because there where Missing data.
import pandas as pd
df = pd.read_csv('https://hfan.greenriverdev.com/SDEV280_project/ClimateData.csv')
df
```

```python
# Next, you can calculate the correlation coefficients between different categories using the corr() function in Pandas. This will give you a correlation matrix for all pairs of numerical columns.

correlation_matrix = df.corr()
correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```
